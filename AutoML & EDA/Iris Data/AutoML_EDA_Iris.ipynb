<b>Importing Libraries

```python
import pandas as pd 
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from supervised.automl import AutoML
from supervised.preprocessing.eda import EDA
```

<b>Loading And Preparing Iris Dataset

```python
iris = load_iris(as_frame=True) 
```

```python
X = iris.data
```

```python
X.shape
```

```python
Y = iris.target
```

```python
Y.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, train_size=0.8, random_state=10)
```

<b>Performing Automated EDA 

```python
EDA.extensive_eda(X_train, Y_train, save_path="/sample/")
```

<b>Performing AutoML Processes

```python
a = AutoML(algorithms=["Xgboost","CatBoost","Random Forest"], mode="Compete", total_time_limit=10)
```

```python
a.fit(X_train,Y_train)
```

<b>Generating AutoML Report

```python
a.report()
```
